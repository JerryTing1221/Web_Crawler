```python
! pip install selenium

import os
import requests
from bs4 import BeautifulSoup
import ssl
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

ssl._create_default_https_context = ssl._create_unverified_context


url = 'https://www.104.com.tw/jobs/search/?ro=0&order=11&asc=0&page={}&mode=s&jobsource=2018indexpoc'
userAgent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'

headers = {
    'user-agent': userAgent
}



page = 0


JobList = []

for i in range(0,50): 
    
    res = requests.get(url=url.format(page), headers=headers)
    html = res.text
    #print(html.text)

    soup = BeautifulSoup(html, 'html.parser')

    
    pageLink = soup.select('a[class="js-job-link"]')


    for p in pageLink: 
        Link = 'https:' + p['href']

        driver = webdriver.Chrome('./chromedriver')
        
        
        

        driver.get(Link)

        
        #職位名稱
        Title = driver.find_element_by_class_name("pr-6.text-break")
        workTitle = Title.text
        #print('職位名稱: ')
        #print(workTitle)
        #print('\n')

        #網址
        #print('網址: ')
        #print(Link)
        #print('\n')
        
        #公司名稱
        Company = driver.find_element_by_class_name("btn-link.t3.mr-6")
        workCompany = Company.text
        #print('公司名稱: ')
        #print(workCompany)
        #print('\n')
   
        #工作內容
        Content = driver.find_element_by_class_name("mb-5.r3.job-description__content.text-break")
        workContent = Content.text
        #print('工作內容: ')
        #print(workContent)
        #print('\n')

        
        
        
        
        elements = driver.find_elements_by_class_name("row.mb-2")
        #0.職業類型 1.工作待遇 2.工作性質 3.上班地點 4.管理責任 5.出差外派 6.上班時段 7.休假制度 8.可上班日 9.需求人數 
        #10.接受身份 11.工作經歷 12.學歷要求 13.科系要求 14.語文條件 15.擅長工具 16.工作技能 17.具備駕照 18.其他條件 

        
        for e in elements:
            outline = e.text
            print(outline)
            print('\n')
            
            if '工作待遇' in outline:
                salary = outline.split
#                 print('工作待遇：')
#                 print(salary)
#                 print('\n')
                
            elif '上班地點' in outline:
                location = outline.split
#                 print('上班地點：')
#                 print(location)
#                 print('\n')
                
                
            elif '工作經驗' in outline:
                Exp = outline
#                 print('工作經驗：')
#                 print(Exp)
#                 print('\n')
                
                
            elif '學歷要求' in outline:
                Edu = outline
#                 print('學歷要求：')
#                 print(Edu)
#                 print('\n')
                
                
            elif '科系要求' in outline:
                major = outline
#                 print('科系要求：')
#                 print(major)
#                 print('\n')
                
            elif '語文條件' in outline:
                language = outline
#                 print('語文條件：')
#                 print(language)
#                 print('\n')
                
            elif '擅長工具' in outline:
                tool = outline
#                 print('擅長工具：')
#                 print(tool)
#                 print('\n')
                
            elif '工作技能' in outline:
                skills = outline
#                 print('工作技能：')
#                 print(skills)
#                 print('\n')
                
                
            elif '其他條件' in outline:
                others = outline
#                 print('其他條件：')
#                 print(others)
#                 print('\n')
                
            else:
                pass

        
        
        time.sleep(3)
        driver.close()
        
                  
        
        
        data = {
            '職位名稱':workTitle,
            '網址':Link,
            '公司名稱':workCompany,
            '工作內容':workContent,
            '工作待遇':salary,
            '上班地點':location,
            '工作經驗':Exp,
            '學歷要求':Edu,
            '科系要求':major,
            '語文條件':language,
            '擅長工具':tool,
            '工作技能':skills,
            '其他條件':others
        }
        
        
        
        JobList.append(data)
        
        page+=1
        
        
df = pd.DataFrame(JobList)
        
df.to_excel('test.xlsx')  
```
